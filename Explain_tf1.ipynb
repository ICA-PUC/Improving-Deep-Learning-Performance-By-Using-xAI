```python
%matplotlib inline

import innvestigate
import innvestigate.utils
import keras.applications.vgg16 as vgg16
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.layers import Input, Dense, Conv2D, MaxPool2D, Flatten, Dropout,BatchNormalization,LeakyReLU,GlobalAveragePooling2D
from keras import datasets, layers, models,Model

from keras import applications
import cv2
import numpy as np
import matplotlib.pyplot as plt
from os.path import basename ,join
import os
import glob as glob

###############################
###################
# Code of LRP Explaining AI Tensorflow 1.x
#
# USED at work:
#Improving Deep Learning Performance By Using Explainable Artificial Intelligence (xAI) Approaches
################

####
### Load MOdel:
model_pred=models.load_model('placas/results_placas/generative/resnet_generative.h5',compile=False,custom_objects = {"LeakyReLU":LeakyReLU})
#########################

model_pred.summary()


###################################################################
## If softmax was apply use, if not comment this part ....
model_pred = innvestigate.utils.model_wo_softmax(model_pred)

###############################################################

# Create analyzer
analyzer = innvestigate.create_analyzer('lrp.z', model_pred)

# Another options: 
# 'lrp.z', 'lrp.z_IB', 'lrp.epsilon', 'lrp.epsilon_IB', 'lrp.w_square', 'lrp.flat', 
#'lrp.alpha_beta', 'lrp.alpha_2_beta_1', 'lrp.alpha_2_beta_1_IB', 'lrp.alpha_1_beta_0', 
#'lrp.alpha_1_beta_0_IB', 'lrp.z_plus', 'lrp.z_plus_fast'
#, 'lrp.sequential_preset_a', 'lrp.sequential_preset_b', 'lrp.sequential_preset_a_flat',
#'lrp.sequential_preset_b_flat'outros olhar documentação
##############################################################################


### Load pictures folder:
pasta_raiz_tds_classes=glob.glob('placas/images/*')
for pasta_fotos in pasta_raiz_tds_classes:
    caminho=pasta_fotos
    #print(caminho)
    pasta_fotos=glob.glob(caminho+'/*')
    count=0
    folder_create_classe=basename(caminho)
    classe_folder = os.path.join('placas/results_generative/resnet', folder_create_classe)
    os.mkdir(classe_folder)
   
    for fotos in pasta_fotos:
        print(fotos)
        img=cv2.imread(fotos)
        img=cv2.resize(img,(224,224))


        img_plot=img


        img = np.expand_dims(img, axis=0)


        predict=model_pred.predict(img)



        print(np.argmax(predict))

        # preprocess

        x = applications.mobilenet.preprocess_input(img)



        # Apply analyzer w.r.t. maximum activated output-neuron
        a = analyzer.analyze(x)
        a = np.array(a)
        a_2 = analyzer.analyze(x)
        a_2 = np.array(a_2)


        plt.rcParams["figure.figsize"] = (15,15)

        # Aggregate long color channels and normalize to [-1, 1]
        a = a.sum(axis=np.argmax(np.asarray(a.shape) == 3))
        a /= np.max(np.abs(a))

        a_2 = a_2.sum(axis=np.argmax(np.asarray(a_2.shape) == 3))
        a_2 /= np.max(np.abs(a_2))

        f, axarr = plt.subplots(1,2)
        axarr[0].imshow(img_plot)
        #axarr[0].imshow(img_plot)
        #axarr[1].imshow(a[0][0], cmap="seismic", clim=(-1, 1))
        axarr[1].imshow(a[0], cmap="seismic", clim=(-1,1))
        #axarr[1][0].imshow(img_plot)
        #axarr[1][1].imshow(a_2[0], cmap="seismic", clim=(-1,1))
        #plt.close(img_plot)
        plt.savefig(classe_folder+'/'+str(count)+'.png', bbox_inches = 'tight')
        count=count+1
        plt.clf()
```
